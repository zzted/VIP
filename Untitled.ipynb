```python
import cv2
import numpy as np
import json
import random

def extract_sampled_points(image_path, output_image_path, output_json_path, target_point_count=1200):
    # 1. 读取带有 Alpha 通道的图片
    image = cv2.imread(image_path, cv2.IMREAD_UNCHANGED)
    
    if image is None:
        print("未找到图片，请检查路径。")
        return

    # 2. 提取 Alpha 通道并强制二值化
    alpha_channel = image[:, :, 3]
    _, binary_mask = cv2.threshold(alpha_channel, 127, 255, cv2.THRESH_BINARY)

    # 3. 寻找最外围的拓扑连续轮廓
    contours, _ = cv2.findContours(binary_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

    if not contours:
        print("未能找到任何轮廓。")
        return

    # 4. 获取最大轮廓（猫猫主体）
    largest_contour = max(contours, key=cv2.contourArea)

    # 5. 提取所有轮廓点的 (x, y) 像素坐标
    all_pixel_points = [(point[0][0], point[0][1]) for point in largest_contour]

    # 6. 随机降采样
    if len(all_pixel_points) > target_point_count:
        sampled_pixel_points = random.sample(all_pixel_points, target_point_count)
    else:
        sampled_pixel_points = all_pixel_points

    # 7. 绘制降采样后的预览图
    # 创建一张全黑背景
    preview_img = np.zeros_like(alpha_channel)
    for x, y in sampled_pixel_points:
        # 使用 cv2.circle 将点画上去
        # 这里的半径设为 1，你可以改成 2 或 3 让点在预览图里看起来更大更清晰
        cv2.circle(preview_img, (x, y), 1, 255, -1)

    # 8. 归一化并生成引擎所需的坐标数据 (UE5 格式)
    height, width = alpha_channel.shape
    ue5_formatted_points = []
    
    # 使用 enumerate 获取索引，用于生成唯一的 Name
    for index, (x, y) in enumerate(sampled_pixel_points):
        nx = (x / width) * 2.0 - 1.0
        ny = -((y / height) * 2.0 - 1.0)
        
        # 严格按照 UE5 Data Table 的要求构建字典
        ue5_formatted_points.append({
            "Name": f"Point_{index}",  # 这是解决报错的关键，为每一行提供唯一名称
            "X": nx,                   # 建议大写，匹配 UE5 里的变量名
            "Y": ny                    # 建议大写，匹配 UE5 里的变量名
        })

    # 9. 保存结果 (记得把原本的 normalized_points 换成 ue5_formatted_points)
    cv2.imwrite(output_image_path, preview_img)
    
    with open(output_json_path, 'w') as f:
        json.dump(ue5_formatted_points, f, indent=4)

    # 9. 保存结果
    cv2.imwrite(output_image_path, preview_img)
    
    with open(output_json_path, 'w') as f:
        json.dump(ue5_formatted_points, f, indent=4)
        
    print(f"处理完成！从 {len(all_pixel_points)} 个边缘点中随机抽取了 {len(sampled_pixel_points)} 个点。")
    print(f"点阵预览图已保存至: {output_image_path}")
    print(f"坐标数据已保存至: {output_json_path}")

# 运行脚本（你可以直接在这里修改 1200 这个数值来测试不同的密度）
```

```python
# 运行脚本
extract_sampled_points('mao_3.png', 'cat_points_preview.png', 'points_data.json', target_point_count=800)
```
